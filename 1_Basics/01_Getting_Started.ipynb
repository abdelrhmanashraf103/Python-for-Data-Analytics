## üí¨ Make a Comment

What if you wanted to include something in your code like a note? ü§î

### üéØ How to Comment
- Include a **`#`** before the line
- This is called a **comment**

### ‚ú® Key Points
- Comments **don't execute** anything üö´
- Comments **aren't read** by the Python interpreter when executing your code
- They're just for **human readers** üë•

---

## üìù Comment Examples

### **Single Line Comment**

````python
# This is a comment. Nothing to the right of the "#" sign will be interpreted
````

````python
# This is a comment

print("What's Up, Dude")
print(2 + 2) # This is a comment also
3 * 3
````

### **Multi-Line Comments**


````python
"""
This is a multi-line comment
It can span across multiple lines
Often used for documentation
"""
def my_function():
    return "Hello"

'''
Single quotes also work
This is technically a string, not a comment
but it serves the same purpose when not assigned
'''
x = 10
y = 20
````
